**OBJECTIVE:**
       The model will be given user input which is the "Favourite movie of the user" with that the model should predict 30 movies based on user preference

**Data Source :** https://github.com/ybifoundation/Dataset/raw/main/Movies%20Recommendation.csv

```
import pandas as pd
```

```
movieData = pd.read_csv("https://github.com/ybifoundation/Dataset/raw/main/Movies%20Recommendation.csv")
```

```
featureList=movieData[['Movie_Genre', 'Movie_Language','Movie_Vote','Movie_Keywords','Movie_Tagline', 'Movie_Cast',
       'Movie_Director']].fillna("")
```

```
allMovieList= movieData['Movie_Title'].tolist()
```

```
y=movieData['Movie_Title']
x_text=featureList['Movie_Genre']+" "+featureList["Movie_Language"]+" "+featureList['Movie_Keywords']+" "+featureList['Movie_Tagline']+" "+ str(featureList['Movie_Vote'])+" "+featureList['Movie_Cast']+" "+featureList['Movie_Director']

from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
x=tfid.fit_transform(x_text)
```

```
from sklearn.metrics.pairwise import cosine_similarity
similarityScore= cosine_similarity(x)
```

```
userFavouriteMovie = input("Enter your Favourite Movie :")
import difflib
closestPairMovies = difflib.get_close_matches(userFavouriteMovie,allMovieList)
bestMatch=closestPairMovies[0]
movieIndex = movieData[movieData.Movie_Title==bestMatch]['Movie_ID'].values[0]
```

```
RecommendationMoviesScore=list(enumerate(similarityScore[movieIndex]))
```

```
sortedMoviesList=sorted(RecommendationMoviesScore,key=lambda x:x[1],reverse=True)
```

```
count=0
for movie in sortedMoviesList:
    Index = movie[0]
    print(movieData[movieData.Movie_ID==Index]['Movie_Title'])
    count+=1
    if count==10:
      break
```
